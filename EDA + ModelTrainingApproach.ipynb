```python
def OutlierRemoval(dataset,column_name, minthersh,maxtheresh):
    min_threshold = dataset[column_name].quantile(minthersh)
    max_threshold = dataset[column_name].quantile(maxtheresh)
    return min_threshold,max_threshold   
```

```python
def SettingOutlier(dataset, column_name,MinThershold,MaxThershold ):
    data = dataset[(dataset[column_name] >= MinThershold ) & (dataset[column_name] < MaxThershold)]
    return data
```

```python
min_threshold,max_threshold = OutlierRemoval(df4,'national_inv',0.09,0.99)
print(min_threshold)
print(max_threshold)
df5 = SettingOutlier(df4,'national_inv',min_threshold,max_threshold)
df5.shape
```
